### **(3 pts.) Reproduce figures 12-5, 12-6,..., 12-9 of [1].**

```python
# Basic and needed functions
import numpy as np

def distance(a, b):
    return np.linalg.norm(np.array(a)-np.array(b))
    
def random_point(dim):
    point = [np.random.random() for _ in range(dim)]
    return point

def random_distances(dim, num_pairs):
    distances = np.array([distance(random_point(dim), random_point(dim)) for _ in range(num_pairs)])
    return np.array(distances)
```

```python
dimensions = range(1,101)

avg_distances = []
min_distances = []

np.random.seed(17) # As asked in the test indications

for dim in dimensions:
    distances = random_distances(dim, 10000)
    avg_distances.append(np.mean(distances))
    min_distances.append(np.min(distances))
```

```python
import matplotlib.pyplot as plt
np.random.seed(17)
plt.figure(figsize=(8,6))
plt.plot(dimensions, avg_distances, 'b', label='average distance')
plt.plot(dimensions, min_distances, 'g', label='minimum distance')
plt.xlabel('# of dimensions')
plt.legend()
plt.title('10,000 random distances (Fig. 12.5)')
```

```python
np.random.seed(17)
plt.figure(figsize=(8,6))
plt.plot(dimensions, np.array(min_distances)/np.array(avg_distances), 'b', label='average distance')
plt.xlabel('# of dimensions')
plt.legend()
plt.title('Minimum distances/Average distances (Fig. 12.6)')
```
