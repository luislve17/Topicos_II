# Manipulating Visualization using Matplotlib (Ch. 4) 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

print(style.available)
```

```python
style.use('ggplot')
```

```python
fig, ax = plt.subplots(1)
```

```python
ax.scatter(np.random.random(30), np.random.random(30))
```

```python
fig
```

```python
ax.set_title('Sample plot')
```

```python
fig
```

```python
ax.set_xlabel("$x$")
ax.set_ylabel("$\sigma(x)$")
```

```python
fig
```

```python
x = np.arange(0,1,0.01)
y = x**2
ax.plot(x, y)
fig
```

```python
ax.plot([0,0.4,0.6], [0.2,0.8,0.9], label="test", c='#00ff00')
```

```python
ax.legend()
```

```python
fig
```

```python
import string

values = np.random.randint(4,10,10)
labels = list(string.ascii_uppercase[:10])
```

```python
fig, ax = plt.subplots(1)

ax.bar(labels, values, align='center', color='#95acba')
ax.tick_params(axis='x', rotation=45)
```

```python
fig, ax = plt.subplots(1)

values = np.random.random(5)
labels = list(string.ascii_uppercase[:5])

ax.pie(values, labels = labels, autopct='%.1f%%', explode=(0,0.2,0,0.1,0))
fig
```
